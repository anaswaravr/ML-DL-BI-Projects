```python
import cv2
from tensorflow.keras.models import load_model
from tensorflow.keras.applications.mobilenet_v2 import preprocess_input
import numpy
```

```python
face_cas=cv2.CascadeClassifier('haarcascade_frontalface.xml')
face_cas
```

```python
model=load_model('mask_recog.h5')
```

```python
def face_mask_detector(frame):
    gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    faces=face_cas.detectMultiScale(gray,minNeighbors=4)
    print(faces)
    for (x,y,w,h) in faces:
        face_frame=frame[y:y+h,x:x+w]
        print(face_frame.shape)
        face_frame=cv2.cvtColor(face_frame,cv2.COLOR_BGR2RGB)
        face_frame=cv2.resize(face_frame,(224,224))
        face_frame=face_frame.reshape(1,224,224,3)
        face_frame=preprocess_input(face_frame)
        pred=model.predict(face_frame)[0]
        print(pred)
        mask,without_mask=pred
        label="With Mask" if mask>without_mask else "Without Mask"
        cv2.putText(frame,label,(x,y),cv2.FONT_HERSHEY_COMPLEX,.5,(0,0,255),2)
        cv2.rectangle(frame,(x,y),(x+w,y+h),(255,0,0),2)
    return frame
```

```python
img=cv2.imread("without_mask.jfif")
prediction=face_mask_detector(img)
cv2.imshow("PREDICTION",prediction)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
video=cv2.VideoCapture(0)
while True:
    suc,frame=video.read()
    if not suc:
        break
    frame=face_mask_detector(frame)
    cv2.imshow("Face Mask Detection",frame)
    if cv2.waitKey(1) & 0XFF==ord('a'):
        break
video.release()
cv2.destroyAllWindows()
```
