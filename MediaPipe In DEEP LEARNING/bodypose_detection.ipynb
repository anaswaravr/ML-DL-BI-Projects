```python
import mediapipe as mp
import cv2
```

```python
mp_bodypose=mp.solutions.pose
mp_drawings=mp.solutions.drawing_utils
```

```python
body_pose=mp_bodypose.Pose()
```

```python
video=cv2.VideoCapture(0)
while True:
    suc,img=video.read()
    img=cv2.flip(img,1)
    img1=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    result=body_pose.process(img1)
    print(result.pose_landmarks)
    if result.pose_landmarks:

        mp_drawings.draw_landmarks(
            img,
            result.pose_landmarks,
            mp_bodypose.POSE_CONNECTIONS,
            mp_drawings.DrawingSpec(color=(255,255,0),
            circle_radius=2,thickness=1),
            mp_drawings.DrawingSpec(color=(255,0,0),thickness=1)
            )
    cv2.imshow("DETECT",img)
    if cv2.waitKey(1) & 0XFF==ord('q'):
        break
video.release()
cv2.destroyAllWindows()
```
