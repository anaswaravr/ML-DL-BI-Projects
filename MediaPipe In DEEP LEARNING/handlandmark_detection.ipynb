```python
#pip install mediapipe==0.10.9
import mediapipe as mp
```

```python
#pip install opencv-python
import cv2
```

```python
mphands=mp.solutions.hands
mp_drawings=mp.solutions.drawing_utils
```

```python
hand=mphands.Hands()
```

```python
video=cv2.VideoCapture(0)
while True:
    suc,img=video.read()
    img=cv2.flip(img,1)    #0---> flipping based on x-axis   1---> flipping based on y-axis
    img1=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    result=hand.process(img1)
    print(result.multi_hand_landmarks)
    if result.multi_hand_landmarks:
        for landmarks in result.multi_hand_landmarks:
            mp_drawings.draw_landmarks(img,landmarks,mphands.HAND_CONNECTIONS,
                                       mp_drawings.DrawingSpec(color=(255,255,0),
                                       circle_radius=4,thickness=2),
                                       mp_drawings.DrawingSpec(color=(255,0,0),thickness=2))

    cv2.imshow("WEBCAM",img)
    if cv2.waitKey(1) & 0XFF==ord('q'):
        break
video.release()
cv2.destroyAllWindows()
```
