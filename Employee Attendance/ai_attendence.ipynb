```python
import cv2
import os
import numpy as np
import face_recognition
```

```python
path='images'
```

```python
mylist=os.listdir(path)
mylist
```

```python
imgs=[]
classnames=[]
for img in mylist:
    imgpath=os.path.join(path,img)
    #print(imgpath)
    curr_img=cv2.imread(imgpath)
    #print(curr_img)
    imgs.append(curr_img)
    #classnames.append(img[:-4])
    classnames.append(img.split(".")[0])
print(classnames)
```

```python
def faceencoding(images):
    encoded_list=[]
    for img in images:
        img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
        face_in_frame=face_recognition.face_locations(img)
        encode=face_recognition.face_encodings(img,face_in_frame)[0]
        #print(encode)
        encoded_list.append(encode)
    return encoded_list
```

```python
encoded_list_knownfaces=faceencoding(imgs)
print(len(encoded_list_knownfaces))
```

```python
import csv
import os
from datetime import datetime

def mark_attendance(name):

    now = datetime.now()
    date_string = now.strftime("%Y-%m-%d")
    time_string = now.strftime("%H:%M:%S")

    file_exists = os.path.exists("attendance.csv")

    # Create file with header if it doesn't exist
    if not file_exists:
        with open("attendance.csv","w",newline="") as f:
            writer = csv.writer(f)
            writer.writerow(["Name","Date","Time"])

    already_marked = False

    with open("attendance.csv","r") as f:
        reader = csv.reader(f)

        for row in reader:
            if len(row) > 0:
                if row[0] == name and row[1] == date_string:
                    already_marked = True
                    break

    if not already_marked:
        with open("attendance.csv","a",newline="") as f:
            writer = csv.writer(f)
            writer.writerow([name,date_string,time_string])
```

```python
import winsound
video=cv2.VideoCapture(0)
while True:
    suc,img=video.read()
    img1=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    face_in_frame=face_recognition.face_locations(img1)
    #print(face_in_frame)
    encode=face_recognition.face_encodings(img1,face_in_frame)
    #print(encode)
    lst=[]
    for loc,enc in zip(face_in_frame,encode):
        matches=face_recognition.compare_faces(encoded_list_knownfaces,enc)
        #print(matches)
        distance=face_recognition.face_distance(encoded_list_knownfaces,enc)
        #print(distance)
        match_index=np.argmin(distance)
        #print(match_index)
        if matches[match_index]==True:
            name=classnames[match_index]
            print(name)
            y1,x2,y2,x1=loc
            cv2.rectangle(img,(x1,y1),(x2,y2),(0,255,255),3)
            cv2.putText(img,name,(x1,y1),cv2.FONT_HERSHEY_COMPLEX,2,(255,0,0),2)
            mark_attendance(name)
        else:
            cv2.rectangle(img,(x1,y1),(x2,y2),(0,255,255),3)
            cv2.putText(img,"UNKNOWN",(x1,y1),cv2.FONT_HERSHEY_COMPLEX,2,(255,0,0),2)
            winsound.Beep(1000, 500)
    cv2.imshow("WEBCAM",img)
    if cv2.waitKey(1) & 0XFF==ord('q'):
        break
video.release()
cv2.destroyAllWindows()
```
